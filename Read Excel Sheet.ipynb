```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans
random_state = 170
xlsx = pd.ExcelFile("16_sa05_cy_B0110O14_315a_01181115.xls")
df = pd.read_excel(xlsx, "Statistics_1-016", index_col=0)
```

```python
df["mAh"] = df["Discharge_Capacity(Ah)"]*1000
da = df[["mAh","Internal_Resistance(Ohm)"]]
```

```python
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(da)
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(da["mAh"], da["Internal_Resistance(Ohm)"], c=y_pred)
```
