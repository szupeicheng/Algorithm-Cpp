```python
import numpy as np
from scipy import linalg
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import data
from skimage import exposure
from skimage import filters
from skimage.color import rgb2hsv
from skimage.exposure import match_histograms

```

```python
rgb_img = mpimg.imread('cloud.jpg')
img = mpimg.imread('city.jpg')
plt.imshow(rgb_img)
```

```python
hsv_img = rgb2hsv(rgb_img)
hue_img = hsv_img[:, :, 0]
value_img = hsv_img[:, :, 2]
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(8, 2))

ax0.imshow(rgb_img)
ax0.set_title("RGB image")
ax0.axis('off')
ax1.imshow(hue_img, cmap='hsv')
ax1.set_title("Hue channel")
ax1.axis('off')
ax2.imshow(value_img)
ax2.set_title("Value channel")
ax2.axis('off')

fig.tight_layout()
```

```python
mpimg.imsave('cloud_rgb.jpg', rgb_img)
```

```python
mpimg.imsave('cloud_hue.jpg', hue_img)
mpimg.imsave('cloud_value.jpg', value_img)
```

```python
value_threshold = 0.3
binary_img = value_img < value_threshold
plt.imshow(binary_img)
mpimg.imsave('cloud_binary.jpg', binary_img)
```

```python
ref_hist, bins = exposure.histogram(rgb_img[:, :, 2], source_range='dtype')
plt.plot(bins, ref_hist / ref_hist.max())
ref_cdf, bins = exposure.cumulative_distribution(rgb_img[:, :, 2])
plt.plot(bins, ref_cdf)
```

```python
ref_hist, bins = exposure.histogram(hsv_img[:, :, 2], source_range='dtype')
plt.plot(bins, ref_hist / ref_hist.max())
ref_cdf, bins = exposure.cumulative_distribution(hsv_img[:, :, 2])
plt.plot(bins, ref_cdf)
```
